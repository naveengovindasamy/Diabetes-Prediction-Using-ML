**(2). Data Preprocessing**

(2)(i).Data Cleaning

```python

#Some features contain 0, so I replaced 0 value by NaN:
df[['Glucose','BloodPressure','SkinThickness','Insulin','BMI']] = df[['Glucose','BloodPressure','SkinThickness','Insulin','BMI']].replace(0,np.NaN)
```

```python
df.head()
```

```python
#Dealing with Missing values
df.isnull().sum()
```

```python
#plotting the missing values

import missingno as msno
msno.bar(df);
```

```python
#Dealing the missing values with the median values. Median values will be applied to missing values in each of the variable
def median_target(var):   
    temp = df[df[var].notnull()]
    temp = temp[[var, 'Outcome']].groupby(['Outcome'])[[var]].median().reset_index()
    return temp
```

```python
columns = df.columns
columns = columns.drop("Outcome")
for i in columns:
    median_target(i)
    df.loc[(df['Outcome'] == 0 ) & (df[i].isnull()), i] = median_target(i)[i][0]
    df.loc[(df['Outcome'] == 1 ) & (df[i].isnull()), i] = median_target(i)[i][1]
```

**(2).(ii).Outlier Analysis**

```python
#To check outliers
for feature in df:
    
    Q1 = df[feature].quantile(0.25)
    Q3 = df[feature].quantile(0.75)
    IQR = Q3-Q1
    lower = Q1- 1.5*IQR
    upper = Q3 + 1.5*IQR
    
    if df[(df[feature] > upper)].any(axis=None):
        print(feature,"yes")
    else:
        print(feature, "no")
```

```python
#Boxplot of Insulin variable
#import seaborn as sns
sns.boxplot(x = df["Insulin"]);
```

```python
#Suppressing the contradictory values for insulin variable alone
Q1 = df.Insulin.quantile(0.25)
Q3 = df.Insulin.quantile(0.75)
IQR = Q3-Q1
lower = Q1 - 1.5*IQR
upper = Q3 + 1.5*IQR
df.loc[df["Insulin"] > upper,"Insulin"] = upper
```

**(3).Encoding**

```python
#Transforming newly created categorical variables to numerical variables
df = pd.get_dummies(df, columns =["NewBMI","NewInsulinScore", "NewGlucose"], drop_first = True)
```

```python
df.head()
```

```python
categorical_df = df[['NewBMI_Obesity i','NewBMI_Obesity ii', 'NewBMI_Obesity iii', 'NewBMI_Overweight','NewBMI_Underweight',
                     'NewInsulinScore_Normal','NewGlucose_Low','NewGlucose_Normal', 'NewGlucose_Overweight', 'NewGlucose_Secret']]
```

```python
categorical_df.head()
```

```python
y = df["Outcome"]
X = df.drop(["Outcome",'NewBMI_Obesity i','NewBMI_Obesity ii', 'NewBMI_Obesity iii', 'NewBMI_Overweight','NewBMI_Underweight',
                     'NewInsulinScore_Normal','NewGlucose_Low','NewGlucose_Normal', 'NewGlucose_Overweight', 'NewGlucose_Secret'], axis = 1)
cols = X.columns
index = X.index
```

```python
X.head()
```

```python
#To increase the model performance we need to do standardization
from sklearn.preprocessing import RobustScaler
transformer = RobustScaler().fit(X)
X = transformer.transform(X)
X = pd.DataFrame(X, columns = cols, index = index)
```

```python
X.head()
```

```python
X = pd.concat([X,categorical_df], axis = 1)
```

```python
X.head()
```

***(4).Creating the base model***


```python
models = []
models.append(('LR', LogisticRegression(random_state = 12345)))
models.append(('KNN', KNeighborsClassifier()))
models.append(('CART', DecisionTreeClassifier(random_state = 12345)))
models.append(('RF', RandomForestClassifier(random_state = 12345)))
models.append(('SVM', SVC(gamma='auto', random_state = 12345)))
models.append(('XGB', GradientBoostingClassifier(random_state = 12345)))
models.append(("LightGBM", LGBMClassifier(random_state = 12345)))
```

```python
#Evaluating all the models in turns
results = []
names = []
```

```python
for name, model in models:
    
        kfold = KFold(n_splits = 10, random_state = 12345, shuffle = True)
        cv_results = cross_val_score(model, X, y, cv = 10, scoring= "accuracy")
        results.append(cv_results)
        names.append(name)
        msg = "%s: %f (%f)" % (name, cv_results.mean(), cv_results.std())
        print(msg)

#Boxplot alogorithm comparison
fig = plt.figure(figsize=(15,10))
fig.suptitle('Algorithm Comparison')
ax = fig.add_subplot(111)
plt.boxplot(results)
ax.set_xticklabels(names)
plt.show()
```
